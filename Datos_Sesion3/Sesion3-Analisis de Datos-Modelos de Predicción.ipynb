### 3. Preparación de los datos

```python
df
```

```python
# Columnas del dataset
df.columns
```

```python
# Solo considerar las columnas SEXO 	EDAD 	CANT_ATENCIONES_CEX
df = df.drop(columns=['   ', 'UUID', 'HISTORIA', 'FEC_FILIACION', 'UBIGEO_LUGAR_RESIDENCIA', 'LUGAR_RESIDENCIA', 'FECHA_CORTE', 'DPTO'])
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Convertir los valores de la columna SEXO en
# FEMENINO = 0, MASCULINO = 1
le_sexo = LabelEncoder()
df['SEXO'] = le_sexo.fit_transform(df['SEXO'])
```

```python
df['SEXO']
```

```python
df
```

```python
# X es la variable de características
# y es la variable objetivo
X = df.drop(columns=['CANT_ATENCIONES_CEX'])
y = df['CANT_ATENCIONES_CEX']
X
```

```python
y
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
 X_test
```

```python
# Escanldar los datos
scaler = StandardScaler()
```

```python
# Actualizar los valores estandarizados
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

### 4. Aplicación de Modelos Predictivos

```python
# Regresión lineal
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Entrenar el modelo de Regresión Lineal
lr_model = LinearRegression()
lr_model.fit(X_train, y_train) # Hallar los parámetros que permiten predicciones
```

```python
# Predecir y evaluar
y_pred_lr = lr_model.predict(X_test)
print('Regresión Lineal:')
print('MAE:', mean_absolute_error(y_test, y_pred_lr))
print('MSE:', mean_squared_error(y_test, y_pred_lr))
print('RMSE:', mean_squared_error(y_test, y_pred_lr, squared=False))
print('R^2:', r2_score(y_test, y_pred_lr))
```

```python
# Arbol de decisiones
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Entrenar el modelo de Árbol de Decisión
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
# Predecir y evaluar
y_pred_dt = dt_model.predict(X_test)
print('Árbol de Decisión:')
print('MAE:', mean_absolute_error(y_test, y_pred_dt))
print('MSE:', mean_squared_error(y_test, y_pred_dt))
print('RMSE:', mean_squared_error(y_test, y_pred_dt, squared=False))
print('R^2:', r2_score(y_test, y_pred_dt))
```
